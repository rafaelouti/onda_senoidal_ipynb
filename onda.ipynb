```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Configuração do gráfico
fig, ax = plt.subplots()
x = np.linspace(0, 2 * np.pi, 100)
line, = ax.plot(x, np.sin(x), lw=2)
```

```python
# Ajuste dos eixos
ax.set_ylim(-1.2, 1.2)
ax.set_xlim(0, 2 * np.pi)
ax.set_title("Onda Senoidal com Cores Dinâmicas")
ax.set_xlabel("X")
ax.set_ylabel("Seno(X)")
```

```python
# Lista de cores vibrantes
colors = ["red", "blue", "green", "purple", "orange", "cyan", "magenta", "yellow"]

# Função de atualização da animação
def update(frame):
    line.set_ydata(np.sin(x + frame / 10.0))  # Move a onda
    line.set_color(colors[frame % len(colors)])  # Muda a cor da linha
    return line,
```

```python
# Criar a animação
ani = animation.FuncAnimation(fig, update, frames=100, interval=50, blit=True)

# Exibir no Jupyter Notebook
from IPython.display import HTML
HTML(ani.to_jshtml())
```
